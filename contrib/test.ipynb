```python
replay = Replay()
```

```python
import gym
```

```python
env = gym.make('CartPole-v0')
```

```python
state = env.reset()
for _ in range(10000):
    action = env.action_space.sample()
    next_state, reward, done, info = env.step(action)
    replay.push(
        torch.Tensor(state),
        torch.tensor(action),
        torch.Tensor(next_state),
        torch.tensor(reward),
        torch.tensor(done)
    )
    state = env.reset() if done else next_state
```
